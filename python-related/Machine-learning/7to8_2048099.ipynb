```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.rc('font',family = 'Malgun Gothic')
plt.plot(x,y)

plt.title("나의 첫 그래프")
plt.xlabel("X축")
plt.ylabel("Y축")

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y, color = 'red',label = 'red', linestyle = ":",marker = "*")
plt.plot(y,x,'blue',label="blue",linestyle="--", marker = "+")

plt.title("My first graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend() ##legend는 범례

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

fig = plt.figure(figsize=(12,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x,y,color='red',linestyle=":",marker="*")
ax2.plot(y,x,color='blue',linestyle="-.",marker="+")

ax1.set_title("red :* graph")
ax2.set_title("blue -.+ graph")

plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

plt.plot(x,y,'r:*',label = "red")
plt.title("red :* graph")
# plt.legend()

plt.subplot(1,2,2)
plt.plot(y,x,'b-.+',label="blue")
plt.title("blue -.+ graph")
plt.legend()

plt.show()
```

```python
plt.bar(x,y,color ="red",label="red")
plt.bar(y,x,color = "blue", label = "blue")
plt.title("나의 첫 그래프")
plt.xlabel("X축")
plt.ylabel("Y축")
plt.legend()
plt.show()
```

```python
players = pd.read_csv("Soccerplayers.csv")
players.head()
```

```python
players["나이"].plot(kind="hist", color='orange')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.hist(players["나이"], color='orange')
ax2.hist(players["나이"], color='orange', bins=5)
ax3.hist(players["나이"], color='orange', bins=20)

ax1.set_title("구간설정없음")
ax2.set_title("bin=5")
ax3.set_title("bin=20")
```
